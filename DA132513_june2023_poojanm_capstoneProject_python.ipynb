### Task 2 Identify columns that have null values and perform the null value treatment 

```python
#Check for null values
data.isnull().sum()
```

```python
#Null value treatment
data['tagline'].fillna(data['tagline'].mode()[0],inplace=True)
data['overview'].fillna(data['overview'].mode()[0],inplace=True)
data['release_date'].fillna(data['release_date'].mode()[0],inplace=True)
data['homepage'].fillna(data['homepage'].mode()[0],inplace=True)
data['runtime'].fillna(data['runtime'].mean,inplace=True)
```

```python
#checking for null values
data.isnull().sum()
```

```python
data.info()
```

**Observation**<br>
In task 2 Null Values in dataset is checked & Replaced null Values & Now dataset has no null values

### Task 7 How popularity of movies related with movie budgets?Are they correlated or uncorrelated with each other?

### Task 10: Sort data in descending order based on revenue & filter the top 500 movies.Find measure of central tendency for filtered data: 1. Budget 2.revenue 3.runtime

```python
central_ten=Final_data.sort_values(by='revenue',ascending=False)[:500]
central_ten1=central_ten.loc[:,['budget','revenue','runtime']]
central_ten_dataframe=pd.DataFrame()
central_ten_dataframe['Measures']=['Mean','Median','Mode']

for i in central_ten1.columns:
    central_ten_dataframe[i]=[central_ten1[i].mean(),central_ten1[i].median(),
                             central_ten1[i].mode()[0]]
central_ten_dataframe
```

**Observation**<br>
Mean of budget is 103193736.0 ,revenue is 4.585776e+08 & Runtime is  118.72<br>
Median of budget is 95000000.0 ,revenue is 3.630016e+08 & Runtime is  116.00<br>
Mode of budget is 150000000.0 ,revenue is 2.177643e+08 & Runtime is  115.00<br>

```python
#Box plot for Budget
plt.figure(figsize=(4,3))
sb.boxplot(x=central_ten1['budget'])
plt.title("Boxplot for Budget")
plt.xlabel("Budget")
plt.show()

# Budget field has 2 outliers
```

```python
#Box plot for Revenue
plt.figure(figsize=(4,3))
sb.boxplot(x=central_ten1['revenue'])
plt.title("Boxplot for Revenue")
plt.xlabel("Revenue")
plt.show()

# Revenue field has many outliers
```

```python
#Box plot for Runtime
plt.figure(figsize=(4,3))
sb.boxplot(x=central_ten1['runtime'])
plt.title("Boxplot for Runtime")
plt.xlabel("Runtime")
plt.show()

# Runtime field has many outliers
```
